```python
import os
import torch
import matplotlib.pyplot as plt
from ultralytics import YOLO
import seaborn as sns
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

```

```python
!ls /content/drive/MyDrive/Yolo
```

```python
# Define paths and parameters
data_yaml = '/content/drive/MyDrive/Yolo/dataset.yaml'  # Updated YAML file
weights = 'yolov8n.pt'       # Pre-trained weights (or path to your own weights)
epochs = 120               # Updated number of training epochs
batch_size = 32              # Batch size for training
initial_lr = 0.001           # Initial learning rate (you can adjust this value)

# Check for GPU availability
device = 'cuda' if torch.cuda.is_available() else 'cpu'

# Load the YOLO model
model = YOLO(weights)

# Train the model with the specified parameters
results = model.train(data=data_yaml, epochs=epochs, batch=batch_size,
                      lr0=initial_lr, device=device)
model.save("final120Yolo.pt")
```
